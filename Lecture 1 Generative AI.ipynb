<h2>Machine Learning vs AI</h2>

En general existe una confusion entre los términos machine learning y AI, veamos las diferencias. Por un lado:

<h3>Definition AI:</h3>
AI es una disciplina rama de ciencias computacionales, la teoría y desarrollo de sistemas computacionales capaces de realizar tareas que normalmente requieren inteligencia.

<h3>Definition Machine Learning</h3>
ML es una subcampo de AI, es la programacion o sistemas que entrenan con informacion de entrada y con ello pueden realizar predicciones útiles.

<h2>Types of Machine Learning</h2>

Tenemos dos tipos de modelos de Machine Learning; modelo insupervisado y supervisado

<img src="types_ML.png"  align="middle"/>


Por un lado el modelo supervisado toma informacion de entrada como entrenamiento y realiza un output que luego es comparado con la realidad y de no acordarse con la realidad se programa nuevamente hasta realizar outputs acorde a la realidad. Por ejemplo un restaurante que compara los tips con la cuenta


<img src="ML_supervised_model.png"  align="middle"/>

Por otro lado el modelo insusupervisado toma informacion de entrada como entrenamiento e intenta agrupar la informacion. Por ejemplo los empleados de una compañia donde se agrupan los salarios contra los años en la compañia.
